```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

Searching for the protein

```python
target = new_client.target
target_query = target.search("HIV")
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[4]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = "IC50")
df = pd.DataFrame.from_dict(res)
```

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) > 10000:
    bioactivity_class.append("inactive")
  elif float(i) < 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
mol_cid = []

for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
can_smis = []

for i in df2.canonical_smiles:
  can_smis.append(i)
```

```python
s = ["molecule_chembl_id","canonical_smiles","standard_value"]
df3 = df2[s]
df3 = pd.concat([df3,pd.Series(bioactivity_class)],axis = 1)

df3.rename(columns={0:"bioactivity_class"},inplace=True)
df3
```

```python
# prompt: Using dataframe df3: remove NaN wherever standard_value is NaN

df3 = df3.dropna(subset=['standard_value',"molecule_chembl_id","canonical_smiles","bioactivity_class"])
df3
```

```python
df3.to_csv("bioactivity_preprocessed.csv",index=False)
```
